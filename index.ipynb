# Textual analyses of Malawi 2022 Auditor General Report

```python
from functools import reduce
import nltk
import os
from pdfminer.high_level import extract_text
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Get the names of files in the Malawi folder
malawi_file_names = os.listdir('./Malawi')
```

```python
# Create a dictionary with each filename is matched with the text
malawi_filename_to_text_dict = {}
for name in malawi_file_names:
    malawi_filename_to_text_dict[name] = extract_text(f'./Malawi/{name}').lower()
```

```python
# Lemmatize the texts
malawi_filename_to_tokens = {}
for name in malawi_file_names:
    tokens = nltk.word_tokenize(malawi_filename_to_text_dict[name])
    lemmatizer = nltk.WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    malawi_filename_to_tokens[name] = lemmatized_tokens
```
